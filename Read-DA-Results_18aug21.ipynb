## This script takes all output paths from Ipopt, and - for each path - plots: 1) action level versus beta, and 2) parameter values versus beta.  (2018 august 21).

### To use: 

### 1) After Ipopt has finished, download all output .dat files for all paths to this directory. 

### 2) Create a subdirectory called PATHs/, and copy all .dat files there too.

### 3) Run.

### The output of this script will be, for each path: action-versus-beta, parameters-versus-beta.  Also, within the script itself, you can specify a print-to-screen of any parameter values at whatever value of beta is desired.

```python
import numpy as np
import scipy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math
import os

% matplotlib inline

#####################################################################
### This program:
### 1. Plots logAction(beta) along each path;
### 2. Identifies path with lowest value of logA at final beta;
### 3. Plots parameters(beta) along each path;
### 4. Prints-to-screen the parameter and Action values at the user-specified desired value of beta.

### USER MANUALLY ENTERS THE FOLLOWING. ###
path = 'PATHs/'
D = 27   # number of variables
P = 24 # number of parameters to estimate; (to subtract at end of each row in Ipopt's .dat file)
start = 3  # number of cols to subtract at beginning of .dat file; for Ipopt this is always 3.
beta_val = 27 # final value of beta at which to print-to-screen the parameter values.  
              # CAREFUL: Not all paths necessarily reach the final desired value of beta.  
              # If they did not, in order to loop through all of them without error, you must 
              # specify the highest value of beta that *all* of the paths reached.
#####################################################################

list_logA = []

listing = os.listdir(path)
outfile = open('ParamEsts_AllPaths.txt', 'w')

for infile in listing:
    results = np.loadtxt(infile)
    print "Current file: " + infile
#    print "current file is: " + infile, "length:", len(results)
    filename, file_extension = os.path.splitext(infile)

    print>>outfile, infile

    B = len(results)-1   

    beta = results[:,0]
    Action = results[:,2]
#    print 'beta', beta
#    print 'action', Action
    logA = np.zeros(len(Action))
    
    for i in range (0,len(Action)):
        logA[i] = np.log(Action[i])/np.log(10)

#    print logA[B], 'log Action at final beta'    
    list_logA.append(logA[B])
#    print list_logA
    
    fig0 = plt.figure(0)
    plt.plot(beta,logA,'8')
    plt.ylabel('log Action')
    plt.xlabel('beta')
    fig0.savefig('ActionVBeta_%s' % filename)

    ### PARAMETERS ###

    betaRange = np.arange(0,len(results),1)

    E01 = results[:,-24]    
    E02 = results[:,-23]    
    E10 = results[:,-22]    
    E12 = results[:,-21]    
    E20 = results[:,-20]    
    E21 = results[:,-19]    
    gL0 = results[:,-18]
    gL1 = results[:,-17]
    gL2 = results[:,-16]
    gNa0 = results[:,-15]
    gNa1 = results[:,-14]
    gNa2 = results[:,-13]
    gK0 = results[:,-12]
    gK1 = results[:,-11]
    gK2 = results[:,-10]    
    gCaT0 = results[:,-9]
    gCaT1 = results[:,-8]
    gCaT2 = results[:,-7]    
    p01 = results[:,-6]
    p02 = results[:,-5]
    p10 = results[:,-4]
    p12 = results[:,-3]
    p20 = results[:,-2]
    p21 = results[:,-1]    


    E01str = str(E01)
    E02str = str(E02)
    E10str = str(E10)
    E12str = str(E12)
    E20str = str(E20)
    E21str = str(E21)
    gL0str = str(gL0)
    gL1str = str(gL1)
    gL2str = str(gL2)
    gNa0str = str(gNa0)
    gNa1str = str(gNa1)
    gNa2str = str(gNa2)
    gK0str = str(gK0)
    gK1str = str(gK1)
    gK2str = str(gK2)
    gCaT0str = str(gCaT0)
    gCaT1str = str(gCaT1)    
    gCaT2str = str(gCaT2)    
    p01str = str(p01)
    p02str = str(p02)
    p10str = str(p10)
    p12str = str(p12)
    p20str = str(p20)
    p21str = str(p21)

#    print 'p01= ', results[:,-2]
#    print>>outfile, 'p01= ', p01str 
#    print 'p10= ', results[:,-1]
#    print>>outfile, 'p10= ', p10str 

    beta_val = beta_val
    print 'gIonic VALUES at final BETA value'
    print 'E01= ', results[beta_val,-24], 'E02= ', results[beta_val,-23], 'E10= ', results[beta_val,-22]
    print 'E12= ', results[beta_val,-21], 'E20= ', results[beta_val,-20], 'E21= ', results[beta_val,-19]
    print 'gL0= ', results[beta_val,-18], 'gL1= ', results[beta_val,-17], 'gL2= ', results[beta_val,-16]
    print 'gNa0= ', results[beta_val,-15], 'gNa1= ', results[beta_val,-14], 'gNa2= ', results[beta_val,-13]
    print 'gK0= ', results[beta_val,-12], 'gK1= ', results[beta_val,-11], 'gK2= ', results[beta_val,-10]
    print 'gCaT0= ', results[beta_val,-9], 'gCaT1= ', results[beta_val,-8], 'gCaT2= ', results[beta_val,-7]
    print 'p01= ', results[beta_val,-6], 'p02= ', results[beta_val,-5], 'p10= ', results[beta_val,-4]
    print 'p12= ', results[beta_val,-3], 'p20= ', results[beta_val,-2], 'p21= ', results[beta_val,-1]    
    
    start = 3  # number of cols to subtract at beginning; for IPOPT this is always 3.
    N = (len(results[0,:]) - start - P)/D   # number of timepoints per time series
    
    results2 = np.zeros((D,N))
    i = 0
    while i < D:
        j = 0
        while j < N:
            results2[i,j] = results[beta_val,(3+i)+D*j]
            j = j+1    
        i = i+1

    P1 = results2[0,:]
    P2 = results2[1,:]
    P3 = results2[2,:]
    P4 = results2[3,:]
    P5 = results2[4,:]
    P6 = results2[5,:]
    time = results2[6,:]

    newICs = np.zeros(D)
    p = 0
    while p < D:
        newICs[p] = results2[p,N-1]
        p = p+1

    print newICs, 'New ICs'

    fig1, ax1 = plt.subplots(3,3)
    ax1[0,0].plot(betaRange,gL0,'8')
    ax1[0,0].axhline(y=0.3, linewidth=2, color = 'r')
    ax1[0,0].set_title('gL0')
    ax1[0,1].plot(betaRange,gNa0,'8')
    ax1[0,1].axhline(y=1.2, linewidth=2, color = 'r')
    ax1[0,1].set_title('gNa0')
    ax1[0,2].plot(betaRange,gK0,'8')
    ax1[0,2].axhline(y=0.2, linewidth=2, color = 'r')
    ax1[0,2].set_title('gK0')    
    ax1[1,0].plot(betaRange,gL1,'8')
    ax1[1,0].axhline(y=0.33, linewidth=2, color = 'r')
    ax1[1,0].set_title('gL1')
    ax1[1,1].plot(betaRange,gNa1,'8')
    ax1[1,1].axhline(y=1.0, linewidth=2, color = 'r')
    ax1[1,1].set_title("gNa1")
    ax1[1,2].plot(betaRange,gK1,'8')
    ax1[1,2].axhline(y=0.22, linewidth=2, color = 'r')
    ax1[1,2].set_title("gK1")
    ax1[2,0].plot(betaRange,gL2,'8')
    ax1[2,0].axhline(y=0.29, linewidth=2, color = 'r')    
    ax1[2,0].set_title("gL2")  
    ax1[2,1].plot(betaRange,gNa2,'8')
    ax1[2,1].axhline(y=1.4, linewidth=2, color = 'r')
    ax1[2,1].set_title("gNa2")       
    ax1[2,2].plot(betaRange,gK2,'8')
    ax1[2,2].axhline(y=0.17, linewidth=2, color = 'r')
    ax1[2,2].set_title("gK2")       
    plt.show()
    fig1.savefig('Params_gLgNagK_%s' % filename)

    fig2, ax2 = plt.subplots(3,2)
#    ax2[0,0].plot(betaRange,gKCa0,'8')
    ax2[0,0].axhline(y=0.24, linewidth=2, color = 'r')
    ax2[0,0].set_title('gKCa0')
    ax2[0,1].plot(betaRange,gCaT0,'8')
    ax2[0,1].axhline(y=0.1, linewidth=2, color = 'r')
    ax2[0,1].set_title('gCaT0')
#    ax2[1,0].plot(betaRange,gKCa1,'8')
    ax2[1,0].axhline(y=0.2401, linewidth=2, color = 'r')
    ax2[1,0].set_title('gKCa1')
    ax2[1,1].plot(betaRange,gCaT1,'8')
    ax2[1,1].axhline(y=0.11, linewidth=2, color = 'r')
    ax2[1,1].set_title('gCaT1')    
#    ax2[2,0].plot(betaRange,gKCa2,'8')
    ax2[2,0].axhline(y=0.2399, linewidth=2, color = 'r')
    ax2[2,0].set_title('gKCa2')
    ax2[2,1].plot(betaRange,gCaT2,'8')
    ax2[2,1].axhline(y=0.09, linewidth=2, color = 'r')
    ax2[2,1].set_title('gCaT2')
    plt.show()
    fig2.savefig('Params_gKCagCaL_%s' % filename)

    fig3, ax3 = plt.subplots(2,3)
    ax3[0,0].plot(betaRange,p01,'8')
    ax3[0,0].axhline(y=0.25, linewidth=2, color = 'r')
    ax3[0,0].set_title('p01')    
    ax3[0,1].plot(betaRange,p10,'8')
    ax3[0,1].axhline(y=0.28, linewidth=2, color = 'r')
    ax3[0,1].set_title('p10')  
    ax3[0,2].plot(betaRange,p20,'8')
    ax3[0,2].axhline(y=0.21, linewidth=2, color = 'r')
    ax3[0,2].set_title('p20')        
    ax3[1,0].plot(betaRange,p02,'8')
    ax3[1,0].axhline(y=0.4, linewidth=2, color = 'r')
    ax3[1,0].set_title('p02')
    ax3[1,1].plot(betaRange,p12,'8')
    ax3[1,1].axhline(y=0.18, linewidth=2, color = 'r')
    ax3[1,1].set_title('p12')        
    ax3[1,2].plot(betaRange,p21,'8')
    ax3[1,2].axhline(y=0.32, linewidth=2, color = 'r')
    ax3[1,2].set_title('p21')        
    plt.show()
    fig3.savefig('Params_gij_%s' % filename)
    
    fig4, ax4 = plt.subplots(2,3)
    ax4[0,0].plot(betaRange,E01,'8')
    ax4[0,0].axhline(y=-83.0, linewidth=2, color = 'r')
    ax4[0,0].set_title('E01')    
    ax4[0,1].plot(betaRange,E02,'8')
    ax4[0,1].axhline(y=-83.3, linewidth=2, color = 'r')
    ax4[0,1].set_title('E02')  
    ax4[0,2].plot(betaRange,E10,'8')
    ax4[0,2].axhline(y=-82.7, linewidth=2, color = 'r')
    ax4[0,2].set_title('E10')        
    ax4[1,0].plot(betaRange,E12,'8')
    ax4[1,0].axhline(y=-82.5, linewidth=2, color = 'r')
    ax4[1,0].set_title('E12')
    ax4[1,1].plot(betaRange,E20,'8')
    ax4[1,1].axhline(y=-83.2, linewidth=2, color = 'r')
    ax4[1,1].set_title('E20')        
    ax4[1,2].plot(betaRange,E21,'8')
    ax4[1,2].axhline(y=-82.9, linewidth=2, color = 'r')
    ax4[1,2].set_title('E21')        
    plt.show()
    fig4.savefig('Params_Eij_%s' % filename)
    
min_logA = min(list_logA)

print "min logA:", min_logA, "   Path number:", list_logA.index(min(list_logA)) + 1      
#fig1.savefig("A_AllPaths.png")    
#fig2.savefig("Params_AllPaths.png")
```
